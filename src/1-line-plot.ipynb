```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.arange(0, 11)

# Calculate y values (square of x)
y = x**2

# Create the line plot
plt.plot(x, y)

# Add title
plt.title("Square of Numbers")

# Add labels for x and y axes
plt.xlabel("X-axis (Values from 0 to 10)")
plt.ylabel("Y-axis (Square of X)")

# Add a grid
plt.grid(True)

# Display the plot
plt.show()
```
