```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
# Line plot data
x_line = np.arange(0, 10)
y_line = x_line * 2

# Bar plot data
categories_bar = ['X', 'Y', 'Z']
values_bar = [10, 15, 7]

# Scatter plot data
x_scatter = [1, 3, 5, 7, 9]
y_scatter = [2, 8, 1, 6, 3]

# Pie chart data
labels_pie = ['Part 1', 'Part 2', 'Part 3']
sizes_pie = [30, 45, 25]

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2)

# Plot in the first subplot (top-left)
axes[0, 0].plot(x_line, y_line)
axes[0, 0].set_title('Line Chart')

# Plot in the second subplot (top-right)
axes[0, 1].bar(categories_bar, values_bar)
axes[0, 1].set_title('Bar Chart')

# Plot in the third subplot (bottom-left)
axes[1, 0].scatter(x_scatter, y_scatter)
axes[1, 0].set_title('Scatter Plot')

# Plot in the fourth subplot (bottom-right)
axes[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%')
axes[1, 1].set_title('Pie Chart')

# Adjust layout to prevent overlapping titles
plt.tight_layout()

# Display the plots
plt.show()
```
